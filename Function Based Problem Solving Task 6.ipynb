```python
#1.Smart Parking Lot Management System
def smart_parking_system(capacity, logs):
    parked = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1
        
        # Track peak usage
        if parked > peak:
            peak = parked
    
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    
    # Check capacity alert
    if parked > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
#2.Online Food Delivery Time Estimator
def estimate_delivery(distance, traffic, weather):
    # Base time: 5 minutes per km
    time = distance * 5
    
    # Traffic delay
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    
    # Weather delay
    if weather == "Rainy":
        time += 5
    elif weather == "Stormy":
        time += 15
    
    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery(8, "High", "Rainy")
```

```python
#3.Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")


# Sample Input (150 booked seats out of 200)
seat_occupancy(200, [1]*150)
```

```python
#4.Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    
    print("Average CPU Load:", int(avg), "%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
#5.Smart Classroom Resource Usage Monitor
def resource_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:   # Overuse threshold (8 hours)
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage_data)
```

```python
#6.Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    confirmed = 0
    waitlist = 0

    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", "Closed" if confirmed == capacity else "Open")


# Sample Input
event_registration(100, 105)
```
