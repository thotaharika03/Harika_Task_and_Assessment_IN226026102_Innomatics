```python
#1.Unique Words in a Sentence
sentence=input("enter a sentence:")
words=sentence.split()
unique_words=set(words)
print("unique words count:",len(unique_words))
print("unique words:",unique_words)
    
```

```python
#2.Highest Salary from Employee Data
employees = {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
highest_emp=max(employees,key=employees.get)    #employees.get tells max() to compare dictionary values (salaries)
highest_salary=employees[highest_emp]            #Access Salary
print("Highest Salary:",highest_emp,"-",highest_salary)
```

```python
#3.Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
print("List:",numbers)
print("Max:",max(numbers))
print("Min:",min(numbers))
```

```python
#4.Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
product_count=0
for i in prices:
    if i>=1000:
        product_count+=1
print("product above 1000:",product_count)
```

```python
#5.Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
# Count Present days
present_count = attendance.count("P")
# Total days
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
#6.Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
Unique_phone_numbers=set(phone_numbers)
print("Unique Phone Numbers:",Unique_phone_numbers)
```

```python
#7.Count Character Frequency
text="python"
char_count={}
for ch in text:
    if ch in char_count:
        char_count[ch]+=1  #If character already exists in dictionary : increment count
    else:
        char_count[ch]=1    #add it with count 1   
print(char_count)
```
